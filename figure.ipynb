```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 4))

# Create boxes for the encoder and decoder
encoder_box = plt.Rectangle((0.2, 0.2), 0.6, 0.6, fill=False, edgecolor='b', linewidth=2, label='Encoder')
decoder_box = plt.Rectangle((1.2, 0.2), 0.6, 0.6, fill=False, edgecolor='g', linewidth=2, label='Decoder')

# Add the boxes to the plot
ax.add_patch(encoder_box)
ax.add_patch(decoder_box)

# Create arrows to represent the flow
arrow = plt.arrow(0.8, 0.5, 0.4, 0, head_width=0.05, head_length=0.05, fc='k', ec='k')

# Add the arrow to the plot
ax.add_patch(arrow)

# Set axis limits and labels
ax.set_xlim(0, 2)
ax.set_ylim(0, 1)
ax.set_xticks([])
ax.set_yticks([])

# Add a legend
ax.legend()

# Add a title
ax.set_title('Autoencoder Architecture')

# Show the plot
plt.show()
```
